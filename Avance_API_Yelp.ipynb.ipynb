# Transformación

```python
df_yelp= pd.DataFrame(data["businesses"])
display (df_yelp.head())
```

```python
contador= 0
lista_consolidada= []
while contador < 200:
    params={'term':'restaurants','location': ciudad,'limit':50,'offset':contador}
    response=requests.get(api_url,params=params,headers=headers)
    if response.status_code != 200:  
        break   
    
    data=response.json()
    

    df_yelp = pd.json_normalize(
        data['businesses'],
        record_path='categories',
        meta=[
            'id', 'name', 'review_count', 'rating', 'price',
            ['coordinates', 'latitude'],
            ['coordinates', 'longitude'],
            ['location', 'address1']
        ],
        sep='_',
        errors='ignore'
    )
    lista_consolidada.append(df_yelp)
    contador += 50
```

```python
df_api_yelp_v2= pd.concat(lista_consolidada, ignore_index=True)
display(df_api_yelp_v2.head())
```

### Hacemos una ponderación del rating para poder hacer recomendaciones a los usuarios

```python
#(peso de penalización — entre más lejos esté un restaurante de 500 reseñas, más se acerca su rating ponderado 
# al promedio general)

C = df_api_yelp_v2["rating"].mean()
m = 500

v = df_api_yelp_v2["review_count"]
R = df_api_yelp_v2["rating"]

df_api_yelp_v2["rating_ponderado"] = (v / (v + m)) * R + (m / (v + m)) * C
df_api_yelp_v2[["name", "rating", "review_count", "rating_ponderado"]].sort_values("rating_ponderado", ascending=False).head(10)
```

```python
top3_por_categoria= (
    df_api_yelp_v2.sort_values("rating_ponderado", ascending=False)
    .drop_duplicates(subset="name")
    .groupby("categoria_estandarizada").head(3)
    [["name", "categoria_estandarizada", "rating", "review_count", "rating_ponderado", "location_address1"]]
    .sort_values(["categoria_estandarizada","rating_ponderado"],ascending=[True,False])
)
display(top3_por_categoria)
```
